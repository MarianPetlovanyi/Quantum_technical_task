```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

So, we imported everything we need,

Let's load the data we have from csv

```python
features = pd.read_csv('Data/internship_train.csv').iloc[:,:-1]
labels = pd.read_csv('Data/internship_train.csv').iloc[:, -1:]
features, labels
```

Let's convert these 90000 lines in numpy arrays

```python
features = np.array(features)
labels = np.array(labels)
```

We have 52 anonymous fields

Let's get familiar with data
I want to define a simple function that will plot a dependency between X(1 of 52 column) and y(target column)

```python
def plot_func(x, y):
    """
    Plot the values in y against the values in x.
    
    Parameters:
    x  The x values to plot.
    y  The y values to plot.
    """
    fig, ax = plt.subplots()
    ax.scatter(x, y)
    plt.xlabel('Feature Values')
    plt.ylabel('Labels Values')
    plt.title('Plot of Labels vs Features')
    plt.show()
```

Let's call it for every column

```python
for i in range(0,53):
    print(i)#number of column
    plot_func(features[:, i], labels[:])
```

It is easy to see that 51 of 52 column are useless

Only 6th(starting from 0) column shows us that there is a relationship between our data and target column

Let's plot it again

```python
plot_func(features[:,6], labels[:])
```
